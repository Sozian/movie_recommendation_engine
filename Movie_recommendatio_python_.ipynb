```python
import ast
```

```python
def convert(obj):
  L=[];
  for i in ast.literal_eval(obj):
    L.append(i['name']);
  return L;
```

```python
print(m_movies['genres'].apply(convert));
```

```python
m_movies['genres']=m_movies['genres'].apply(convert);
print(m_movies.head());
```

```python
m_movies['keywords']=m_movies['keywords'].apply(convert);
```

```python
def convert3(obj):
  L=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter!=3:
      L.append(i['name'])
      counter+=1
    else:
      break
  return L
```

```python
m_movies['cast']=m_movies['cast'].apply(convert3);
```

```python
print(m_movies.iloc[0].cast);
```

```python
def fetch_directer(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      L.append(i['name']);
  return L;
```

```python
m_movies['crew']=m_movies['crew'].apply(fetch_directer);
```

```python
print(m_movies.head());
```

```python
# overview was given in the string but we need to seperate it by comma
m_movies['overview']=m_movies['overview'].apply(lambda x:x.split())
```

```python
#we are replacing all the spaces from the string
m_movies['genres']=m_movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
m_movies['keywords']=m_movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
m_movies['cast']=m_movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
m_movies['crew']=m_movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
print(m_movies.head());
```

```python
#now we are creating a tag
m_movies['tags']=m_movies['overview']+m_movies['genres']+m_movies['keywords']+m_movies['crew'];
```

```python
print(m_movies.head());
```

```python
new_df=m_movies[['movie_id','title','tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
#converting all the things in lower case
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]

  for i in text.split():
      y.append(ps.stem(i))

  return " ".join(y)
```

```python
new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```
